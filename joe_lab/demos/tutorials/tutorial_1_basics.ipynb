## A Tiny Bit of Background

*joe* generates approximate solutions to evolutionary partial differential equations (PDEs). Currently, *joe* only supports finding solutions depending only on **one space variable** $x$ as well as a single time variable $t$. The spatial domain for our solution $u(x,t)$ is either a closed, symmetric interval with periodic boundary conditions (topologically equivalent to the unit circle $S^1$), or the real line $\mathbb{R}$: in the latter case, we impose the "boundary condition" that $u(x,t)$ vanishes rapidly at spatial infinity. Being able to deal with both of these boundary conditions makes *joe* well-suited to handling reaction-diffusion equations **and** nonlinear dispersive equations. 

*joe* discretizes the spatial derivatives in a PDE using the **fast Fourier transform (FFT)**. If our PDE is first-order in time, this diagonalizes the linear, constant-coefficient part of the PDE in question, which in turn greatly simplifies time-stepping. If we're instead second-order in time, the PDE is not quite diagonalized, but it still has a nice, banded structure: we'll discuss this more in tutorial 2. 

In *joe*, explicit and (formally) fourth-order accurate time-stepping may be performed using the **exponential time differencing Runge-Kutta 4 method (ETDRK4)**. For first-order-in-time equations, *joe* also supports time-stepping with the **integrating factor Runge-Kutta 4 method (IFRK)** or the **exponential Euler method (ETD1)**, but I recommend sticking with ETDRK4 owing both to my own tests and the findings of Kassam & Trefethen 2005 and Klein 2007.

Crucially, using exponential methods means *joe* can solve stiff PDEs without having to resort to a slow implicit method! Additionally, by default *joe* saves all the matrix exponentials and **$\phi$-functions** ($\approx$ exponential Runge-Kutta weights) it computes and loads them when they are required, so the code never does work unless it absolutely needs to. The $\phi$-functions themselves are computed via approximating Cauchy integrals by quadrature as described in Kassam & Trefethen 2005.

Once a time-step in the Fourier domain is completed, we may use the **inverse fast Fourier transform (IFFT)** to obtain our solution in the (physical) spatial domain. In the language of scientific computing, the FFT-based solution strategy in *joe* is called a **spectral method** (strictly speaking, *joe* actually uses a **pseudo-spectral method** since it leverages a spatial-domain computation to assemble nonlinear terms). As long as $u(x,t)$ is smooth, spectral methods can reproduce $u(x,t)$ with high accuracy and great speed (owing to the "fast" in "fast Fourier transform"). So, *joe* uses fast methods to approximate spatial derivatives and fast methods to perform time-stepping. 

Now, in order to compute the FFT of $u(x,t)$ accurately, we have to treat $u$ as if it's a function on the unit circle $S^1$: this is because the FFT is really just the Fourier transform on a uniform lattice approximation of the circle (such a lattice approximation gives rise to a copy of $\mathbb{Z}_{N}$, where $N$ is the number of spatial grid points we sample). Of course, if $u(x,t)$ is subject to periodic boundary conditions this definitely holds and we have nothing to worry about. However, if $u(x,t)$ is supposed to be defined for all $x\in \mathbb{R}$, we need to do a little more work. 

1) First, we pick $\ell\gg 1$ and treat $u(x,t)$ as a function on $\left[-\frac12\ell, \frac12\ell\right] \approx \mathbb{R}$. 

2) We then artificially impose periodic boundary conditions on $u(x,t)$. This works fine as long as $u$ is zero in suitably thick layers about $\pm \frac12\ell$. 

3) To prevent the artificial periodic boundary conditions from causing trouble when our simulations are long enough for $u$ to be *nonzero* near $\pm \frac12\ell$, we introduce a **sponge layer**. This adds a heat-equation-like term to the governing PDE that "turns on" only in a neighbourhood of the boundary: since heat flow rapidly decreases amplitude, the idea is that once the solution leaks into the sponge layer around $x=+\frac12\ell$ it is damped to zero before it can emerge from $x=-\frac12\ell$ and pollute the  approximation. Since heat damping is less effective on low-frequency signals, following Liu & Trogdon 2023 we also add some harsh but infrequent exponential damping in the sponge layer. The part of the solution that gets sent to zero in the sponge layer can be though of as having "gone off to infinity". 


Of course, using sponge layers means that we can only really trust our numerical solution on a smaller subinterval of $\left[-\frac12\ell, \frac12\ell\right]$, say, $\left[-\frac14\ell, \frac14\ell\right]$. Additionally, having to perform a step of heat flow requires a linear solve, which can be a bit expensive. However, the sum cost of these issues is still a small price to pay for the rapidity and accuracy of FFT-based differentiation. 

In summary, *joe* quickly and accurately solves PDEs defined on $S^1$ or $\mathbb{R}$ using three key ingredients: 

1) spectral approximation of derivatives; 

2) fourth-order (explicit!) exponential time-stepping; 

3) the use of sponge layers to let waves "go off to infinity".

### What *joe* Cannot Do, and What it Struggles With 

Now that we understand the basics of how and why *joe* works, I'll briefly go through some of its limitations.

Here are limitations that will be handled in the short term: *joe* cannot...
- solve PDEs for functions having more than one spatial unknown;
- solve PDEs that are order three or greater in time.

Here are the more serious limitations on *joe*'s performance and versatility. 
- *joe* works on PDEs of the form

$$
\partial_{t}^{m}u_{t} +L\left(\frac{1}{i}\partial_{x}\right)u + f(x,t,u,u_{x},u_{t},u_{xx},u_{xt},...) = 0,
$$

where $L$ is a multiplier with *no dependence on $x$* and $f$ is a suitably nice function. So, *joe* just isn't right for equations where all linear terms have variable coefficients.

- *joe* does not support spatial or temporal adaptivity. To my knowledge, even just temporal adaptivity for exponential methods is not yet well-understood! 

- *joe* is based around spectral methods, so it cannot deal with discontinuous solutions such as shock waves. For smooth but rapidly varying solutions, *joe*'s spectral methods may still have trouble owing to the Gibbs phenomenon. 

- While sponge layers do work, they are necessarily parametric (one must choose the computational domain size, sponge layer thickness, damping coefficient, and more). As every applied mathematician knows, parameter tuning can sometimes be a hassle. 

## Code Demo 1: The BBM Solitary Wave on a Periodic Interval

For our first adventure with *joe*, we simulate a solution to the **Benjamin-Bona-Mahony equation (BBM)**

$$
u_{t}-u_{xxt}+u_{xxx} + 6uu_{x} = 0, \quad x\in \mathbb{R}, \ t\geq 0, 
$$

a model of long-wavelength, small-amplitude gravity waves in shallow water. For further physical and mathematical background on this equation, see

- Benjamin et al, Phil. Trans. Roy. Soc. A, 1972, https://www.jstor.org/stable/74079, or
- Albert and Bona, J. Nonlinear Sci., 1991, https://link.springer.com/article/10.1007/BF01238818. 

*A pedantic physics remark: we are studying BBM moving in the frame of the fastest linear waves, so our linear dispersive term looks a bit different than the one the reader may be used to seeing.*

If we denote our Fourier frequency variable by $k$ and the Fourier transform of $u$ by $\widehat{u}$, BBM may be written in Fourier space as

$$
\widehat{u}_{t} + i\frac{k^3}{1+k^2} \widehat{u} -\frac{3ik}{1+k^2}\left(u^2\right)^{\wedge}(k) = 0 . 
$$

For large $|k|$, the $k$-dependence of the linear term is $\mathcal{O}\left(|k|\right)$, so this equation is mildly stiff (*joe* can certainly handle a lot worse!). 

Note that solutions of BBM formally satisfy the mass conservation law 

$$
\int u(x,t) \ \mathrm{d} x = \text{constant}
$$

provided we're on a periodic interval, or $u$ vanishes rapidly at infinity. 

For any wave speed $c\in (-\infty, -1)\cup(0,\infty)$, BBM admits a solitary wave solution of the form

$$
u(x,t) = \frac{c}{2} \text{sech}^2\left(\frac12\sqrt{\frac{c}{1+c}}\left(x-ct\right)\right).
$$

Let's see if *joe* can reproduce this solitary wave solution in  the case $c=2$. We may do this by precribing the initial data 

$$
u(x,0) = \text{sech}^2\left(\frac12\sqrt{\frac23} \ x \right)
$$

and examining the solution that *joe* produces. We can really put *joe* to the test by seeing if the waveform is preserved for long times: in this case, it's OK to use periodic BCs.

### Initializing a Simulation

First, we prescribe what PDE we want to solve. In *joe*, this is done by creating a $\texttt{model}$ (as in "model PDE") object. To input the Fourier-transformed PDE

$$
\partial_{t}^{m} \widehat{u} + L\left(k\right)\widehat{u} + \widehat{f(u)} = 0,
$$

into *joe*, one initializes a $\texttt{model}$ object with the following attributes: 

- a $\texttt{model\_kw}$ string that represents the name of the model
- $\texttt{t\_ord} = m $
- $\texttt{symbol} = L(k)$ (a callable with one argument)
- $\texttt{fourier\_forcing} = \widehat{f(u)}$ (a callable with three arguments $V, k ,x$, **in that order!**)
- a boolean called $\texttt{nonlinear}$ that excludes the nonlinearity in the model when equal to $\texttt{False}$ (by default, $\texttt{nonlinear}=\texttt{True}$).

To concretize this discussion, note that for BBM we have

- $\texttt{model\_kw} = \texttt{`bbm'}$

- $m=1$

- $L(k) = i\frac{k^3}{1+k^2}$

- $\widehat{f} = -\frac{3ik}{1+k^2}\left(u^2\right)^{\wedge}$

- $\texttt{nonlinear} = \texttt{True}$.  

In this case, the forcing has no explicit dependence on the spatial variable $x$ (it depends on $x$ only through $u$). In principle, however, the nonlinear term may look like
$$
f(u,x) = a(x)u^2
$$
for some function $a(x)$, and for this reason the $\texttt{fourier\_forcing}$ callable must take in $x$ as an argument. 

Fortunately, BBM is one of the PDEs that's built into *joe*, so we can just load it up with the following code: 

```python
model_kw = 'bbm' # put in the name of the model equation 
my_model = joe_models.builtin_model(model_kw, nonlinear=True) # load your model object 
```

Alternatively, we could build our $\texttt{model}$ object from scratch using the following script: 

```python
def my_symbol(k):
    return 1j * (k ** 3) / (1. + k ** 2) 

def my_fourier_forcing(V, k, x, nonlinear):
    p = 1.

    out = -6. * float(nonlinear) * (1. / (p + 1.)) * 1j * k / (1. + k ** 2) * rfft(irfft(V) ** (p + 1))
    return out

my_model_custom = joe.model(model_kw, 1, my_symbol, my_fourier_forcing, nonlinear=True)
```

Now that we've put the PDE we want to solve into the computer, we need to specify the initial condition on our solution $u(x,t)$. This is done with an $\texttt{initial\_state}$ object. As with $\texttt{model}$ objects, these can either be called from a module of built-in options...

```python
initial_state_kw = 'bbm_solitary_wave'
my_initial_state = joe_initial_states.builtin_initial_state(initial_state_kw)
```

...or constructed from scratch by the user:

```python
def bbm_solitary_wave(x, c=1.):
    if c <= 0 and c >= -1:
        raise ValueError('BBM soliton speed must be strictly positive, or strictly less than -1.')
    return 0.5*c*(np.cosh(0.5 * np.sqrt(c/(1.+c)) * x) ** -2)

my_initial_state_custom = joe.initial_state('bbm_solitary_wave', lambda x: bbm_solitary_wave(x,c=2.))
```

At this point, we've defined our PDE and initial state, so all that's left is to define our discretization parameters.

```python
T = 50.  # time to stop simulation at

dt = 1e-2  # time step size

nsteps = int(T / dt)  # total num of time steps we take

length = 100. # lower-case l from earlier

# number of grid nodes
N = 2 ** 10

# define the space-time grid as a dict for easy of handling
stgrid = {'length':length, 'T':T, 'N':N, 'dt':dt}
```

Since our solitary wave speed was selected at $c=2$ and the domain has length $\ell=100$, we expect the solitary wave to be exactly where it started when the simulation ends at $T=50$. 

Now, we can define a $\texttt{simulation}$ object by combining the PDE, IC, BCs, and discretization parameters. 

```python
my_sim = joe.simulation(stgrid, my_model, my_initial_state, bc='periodic', ndump=20)
# NOTE 1: space-time grid goes in first, THEN model object, THEN initial_state object, THEN bc.
# NOTE 2: ndump is an optional param telling the solver to only save the sim every 'ndump' steps. 
```

Note that no solve has been done yet: all that's happened is that we've created an object! 

If you want to have a look at the initial conditions, simply call the function $\texttt{plot\_initial\_condition}$ on your simulation object:

```python
my_sim.plot_initial_condition(color='xkcd:cerulean', usetex=True, show_figure=True, save_figure=False)
```

### Running a Simulation

To actually run the solve is easy: just call the $\texttt{run\_sim}$ method on your simulation object!

```python
my_sim.run_sim(method_kw='etdrk4', print_runtime=True) 
```

Note that, by default, we print the wall-clock runtime of the solve. If you don't want this information, simply pass the argument $\texttt{print\_runtime=False}$ into $\texttt{run\_sim}$. 

We can immediately check the quality of the simulation by verifying whether or not it conserved mass to high precision. This can be easily done with the $\texttt{get\_fm}$ function (fm = first moment, which is admittedly an abuse of terminology). Details underlying the numerical integration are provided in tutorial 3. 

```python
# report error in first moment
my_sim.get_fm()
error = np.amax(my_sim.fm_error)
print('Maximum error in mass = %.2E' % error)
```

We see that mass is conserved to machine precision.

### Saving and Loading Simulations

Here is how to save a simulation: 

```python
my_sim.save()
```

The file is stored in a folder called joe_sim_archive instead your current directory: this extra folder is automatically created if it doesn't already exist.  

**WARNING:** saving a $\texttt{simulation}$ object only keeps the $\texttt{model\_kw}$ attribute of the underlying $\texttt{model}$ object. The reason for this admittedly unusual choice has to do with an issue in the "pickle" module that arises from having some flexibility in switching between built-in and custom $\texttt{model}$ objects. So, before you save a simulation, make sure you won't need to compute the symbol or the forcing term from the underlying model!   

This automatically saves the simulation (using the pickle module... for an intro to pickle see https://docs.python.org/3/library/pickle.html) in joe_sim_archive. Again, if the folder isn't there already, the above command will also create it automatically. The reason to use pickle instead of saving the simulation data in a .json file or an .npy file is that pickle easily allows one to save *objects*, not just arrays or dicts. So, when we load an old simulation, we can treat it exactly like a new one: we can extract its attributes, make pictures with it, et cetera: if we just saved the numerical solution as an array in a .npy file, we would lose the benefits of my hard work coding up the $\texttt{simulation}$ class :) Additionally, my own experiments have indicated that only saving *joe* simulations as .npy files rather than .pkl files only decreases their size marginally. 

Once the sim is saved, you can load it as follows: 

```python
my_sim.load()
```

This takes the existing simulation object and populates its solution values (the attribute $\texttt{Udata}$) with loaded data in the form of a numpy array with shape

$$
\text{number of temporal samples} \times \text{number of spatial samples}. 
$$

From here you can treat it just like a simulation you ran. 

In fact, if you've used *joe* lots and forget what simulations you've already done and which ones you haven't, you might want to leverage the lovely function $\texttt{load\_or\_run}$. This looks in $\texttt{sim\_archive}$ for a saved version of the simulation in question and, if found, loads it up. If it doesn't find the simulation in $\texttt{sim\_archive}$, it runs the simulation. Either way, after you execute this function, you will have a $\texttt{simulation}$ object containing simulation data! 

In my opinion, this function should be your default way of performing simulations. 

```python
my_sim.load_or_run(verbose=True, save=True) # verbose means 'tell us if you loaded or ran'
# save=True means save the sim if you had to run it, and save=False means don't save if you run it (helpful for debugging)
```

## Code Demo 2: BBM with a Sponge Layer

Next, let's see how to solve BBM on an approximation of $\mathbb{R}$ rather than a periodic interval. Here, we encounter sponge layers! We'll impose the initial condition 

$$
u(x,0) = 1.3e^{-x^2}
$$

to contrast our results with a similar test case for the Korteweg-de Vries equation (KdV) presented Liu & Trogdon 2023. 

The model PDE has already been defined, and all that needs to change is our initial state object. 

```python
initial_state_kw = 'gaussian_even_alt'
my_initial_state = joe.initial_state(initial_state_kw, lambda x: 1.3*np.exp(-x**2))
```

We'll also use a different space-time grid involving a much larger spatial interval. 

```python
T = 160.  # time to stop simulation at

dt = 1e-2  # time step size

nsteps = int(T / dt)  # total num of time steps we take

length = 400. # lower-case l from earlier

# number of grid nodes
N = 2 ** 11

# define the space-time grid as a dict for easy of handling
stgrid = {'length':length, 'T':T, 'N':N, 'dt':dt}
```

Again, sponge layers are parametric, so we must prescribe those parameters using a $\texttt{dict}$ called $\texttt{sponge\_params}$ that will get passed into the $\texttt{simulation}$ object. 

```python
l_endpt = -length * 0.5 + 0.5 * length * 0.1 # left endpoint for left-side sponge layer
r_endpt = l_endpt + 0.01 * length # right endpoint for right-side sponge layer
width = (2 ** -6) * length / 100. # effective width of transition region
sponge_params = {'l_endpt': l_endpt, 'r_endpt': r_endpt,
                 'width': width, 
                 'expdamp_freq': 100, # how often we use harsh "exponential damping" (every 100 iterations) 
                 'damping_amplitude': 10., # how strong diffusive damping is
                 'splitting_method_kw': 'naive', # splitting method to use for parabolic solve
                 'spongeless_frac': 0.5}  # this is the fraction of the middle of the spatial domain to keep in the plots.
                                          # for example if =0.5 then we throw out the first and last quarter of the grid when plotting
```

We now create the simulation (note how the $\texttt{bc}$ kwarg changes compared to our first demo)...

```python
my_sim = joe.simulation(stgrid, my_model, my_initial_state, bc='sponge_layer', sponge_params=sponge_params, ndump=20)
```

... and we load-or-run the simulation as described above:

```python
my_sim.load_or_run(verbose=True, save=True)
```

Let's have a look at the Hovmoeller plot. 

```python
my_sim.hov_plot(fieldname='u', usetex=True, cmap='cmo.haline', show_figure=False, save_figure=True, dpi=100)
Image('joe_visuals/' + my_sim.picname, embed=True)
```

Notice first that the sponge layer is doing its job: the bright yellow and dark blue lines don't pollute the solution once they "go off to infinity" (hit the sponge layer boundary at $x=\pm 100$). 

On the physical side, observe that the Gaussian initial state evolved into a right-moving solitary wave with positive amplitude (the bright yellow-green line with positive slope), a left-moving solitary wave with negative amplitude (the dark blue line with negative slope) and a left-moving linear wavetrain with short waves at the front and long waves in the back. This suggests that BBM obeys "solitary wave resolution" similar to the "soliton resolution" enjoyed by integrable nonlinear dispersive PDE like KdV. 

Movies are likewise identical to the periodic case (on my computer, this code takes about 27 seconds to run): 

```python
my_sim.save_movie(dpi=80, usetex=False, fieldcolor='xkcd:cerulean', fieldname='u')
```

```python
Video('joe_visuals/' + my_sim.moviename, embed=True)
```

The movie confirms that the sponge layer is working!
